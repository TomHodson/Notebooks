```python
from bs4 import BeautifulSoup as B
import requests
page = requests.get("http://www.primewire.ag/watch-2722947-Adventure-Time-with-Finn-and-Jake-online-free")
print(dir(page))

soup = B(page.text, "html.parser")
for link in soup.find_all('a'):
    if link.span and link.span.has_attr("class"):
        url = link["href"]
        name = link.span.text
        num = link.text[:3]
        print(name, num, url, )
```
